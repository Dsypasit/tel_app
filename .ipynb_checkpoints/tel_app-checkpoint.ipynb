```python
from tkinter import *
from tkinter.ttk import *
from tkinter import messagebox
import json
```

```python
class tel_app:
    
    def __init__(self,root):
        self.w = root
        self.head_list = ['id', 'name', 'tel', 'line id', 'email']
        self.nw = None
        self.run()
        

    def convert_dict(self):
        if '' in [self.idst.get(), self.name.get(), self.tel.get()]:
            messagebox.showerror(title='missing', message='You fill out the missing')
            return False
        else :
            return {'id':self.idst.get(), 
                    'name':self.name.get(),
                    'tel':self.tel.get(),
                    'line_id':self.line_id.get(),
                    'email':self.mail.get()
                   }

    def show_table(self):
        self.ee.delete(0,END)
        try:
            d = json.load(open('tel.json','r'))
        except:
            messagebox.showwarning(title='No data', message='Please add information')
            return False

        self.table.delete(*self.table.get_children())
        
        if not d:
            messagebox.showwarning(title='No data', message='Please add information')

        
        if d != None:
            for i in d:
                data = list(i.values())
                self.table.insert("",'end',values=data)
        else:
            messagebox.showwarning(title='No data', message='Please add information')
            
    def select(self, word):
        self.table.delete(*self.table.get_children())
        d = json.load(open('tel.json', 'r'))
        index_val = self.head_list.index(self.t.get())
        for i in d:
            if word in list(i.values())[index_val]:
                self.table.insert("", 'end', values=list(i.values()))

    def update(self, ids):
        d = json.load(open('tel.json', 'r'))
        new_data = {}
        for c,i in enumerate(d):
            if ids==i['id']:
                if self.convert_dict():
                    d.pop(c)
                    d.append(self.convert_dict())
                else:
                    return False
        json.dump(d, open('tel.json', 'w'))
        self.show_table()
        self.nw.destroy()

    def delete(self):
        d = json.load(open('tel.json', 'r'))
        n = self.table.focus()
        if n:
            data = self.table.item(n)
            for c,i in enumerate(d):
                if data['values'][1] == i['name']:
                    d.pop(c)
            json.dump(d,open('tel.json', 'w'))
            self.table.delete(n)
        
    def add(self):
        try:
            d = json.load(open('tel.json','r'))
        except:
            d=[]
            
        new_data = {}

        if self.convert_dict():
            d.append(self.convert_dict())
            json.dump(d, open('tel.json','w'))
        else:
            return False
        self.show_table()
        self.nw.destroy()
    
    
    def search(self,*args):
        if self.g.get() == '':
            self.show_table()
        else:
            self.select(self.g.get())
            
    def add_tab(self):
        if self.nw == None or not Toplevel.winfo_exists(self.nw):
            self.nw = Toplevel(self.w)
            self.nw.title('Add')

            self.l1 = Label(self.nw, text='id')
            self.l1.grid(row=0, column=0, padx=5, pady=15)

            self.l2 = Label(self.nw, text='name')
            self.l2.grid(row=0, column=2, ipadx=5)

            self.l3 = Label(self.nw, text='tel')
            self.l3.grid(row=0, column=4, ipadx=5)

            self.l4 = Label(self.nw, text='line id')
            self.l4.grid(row=1, column=0)

            self.l5 = Label(self.nw, text='email')
            self.l5.grid(row=1, column=2, ipadx=5)

            self.idst = StringVar()
            self.e1 = Entry(self.nw, textvariable=self.idst)
            self.e1.grid(row=0, column=1, padx=15, pady=15)

            self.name = StringVar()
            self.e2 = Entry(self.nw, textvariable=self.name)
            self.e2.grid(row=0, column=3, padx=15)

            self.tel = StringVar()
            self.e3 = Entry(self.nw, textvariable=self.tel)
            self.e3.grid(row=0, column=5, padx=15)

            self.line_id = StringVar()
            self.e4 = Entry(self.nw, textvariable=self.line_id)
            self.e4.grid(row=1, column=1, padx=15)

            self.mail = StringVar()
            self.e5 = Entry(self.nw, textvariable=self.mail)
            self.e5.grid(row=1, column=3, padx=15, pady=15)

            self.bf = Frame(self.nw)
            self.bf.grid(row=1, column=4, columnspan=2)

            self.bf1 = Button(self.bf,text='Ok', command=self.add)
            self.bf1.grid(row=0, column=0)

            self.bf2 = Button(self.bf,text='cancle', command=self.nw.destroy)
            self.bf2.grid(row=0, column=1)
        
    def update_tab(self):
        if self.nw == None or not Toplevel.winfo_exists(self.nw):
            if self.table.focus():
                self.nw = Toplevel(self.w)
                self.nw.title('Update')
                n = self.table.focus()
                data = self.table.item(n)['values']

                self.l1 = Label(self.nw, text='id')
                self.l1.grid(row=0, column=0, padx=5, pady=15)

                self.l2 = Label(self.nw, text='name')
                self.l2.grid(row=0, column=2, ipadx=5)

                self.l3 = Label(self.nw, text='tel')
                self.l3.grid(row=0, column=4, ipadx=5)

                self.l4 = Label(self.nw, text='line id')
                self.l4.grid(row=1, column=0)

                self.l5 = Label(self.nw, text='email')
                self.l5.grid(row=1, column=2, ipadx=5)

                self.idst = StringVar()
                self.e1 = Entry(self.nw, textvariable=self.idst)
                self.e1.grid(row=0, column=1, padx=15, pady=15)

                self.name = StringVar()
                self.e2 = Entry(self.nw, textvariable=self.name)
                self.e2.grid(row=0, column=3, padx=15)

                self.tel = StringVar()
                self.e3 = Entry(self.nw, textvariable=self.tel)
                self.e3.grid(row=0, column=5, padx=15)

                self.line_id = StringVar()
                self.e4 = Entry(self.nw, textvariable=self.line_id)
                self.e4.grid(row=1, column=1, padx=15)

                self.mail = StringVar()
                self.e5 = Entry(self.nw, textvariable=self.mail)
                self.e5.grid(row=1, column=3, padx=15, pady=15)

                self.bf = Frame(self.nw)
                self.bf.grid(row=1, column=4, columnspan=2)    

                self.e1.insert(END, data[0])
                self.e2.insert(END, data[1])
                self.e3.insert(END, data[2])
                self.e4.insert(END, data[3])
                self.e5.insert(END, data[4])

                _id = self.idst.get()
                self.bf1 = Button(self.bf,text='Ok', command=lambda i=_id: self.update(i))
                self.bf1.grid(row=0, column=0)

                self.bf2 = Button(self.bf,text='cancle', command=self.nw.destroy)
                self.bf2.grid(row=0, column=1)

    def sort_col(self, col, reverse):
        data = [(self.table.set(k, col), k) for k in self.table.get_children()]
        data.sort(key=lambda e:e[0], reverse = reverse)
        
        for index,(n,i) in enumerate(data):
            self.table.move(i, '', index)
        
        self.table.heading(col, text=col, command=lambda _c=col : self.sort_col(_c,not reverse))
        

    def run(self):
        self.w.title('Tel')

        self.head = Frame(self.w)
        self.head.grid(row=0, column=0, columnspan=7, padx=20, pady=15)
        
        #popmenu

        self.t = StringVar(self.w)
        self.t.set(self.head_list[0])
        
        h = self.head_list[:]
        h.insert(0,'id')
        self.popmenu = OptionMenu(self.head,self.t, *h)
        self.ls = Label(self.head, text='select')
        self.ls.grid(row=0, column=0)
        self.popmenu.grid(row=0,column=1)

        self.g = StringVar()
        self.ee = Entry(self.head, textvariable=self.g, width=40, justify='center')
        self.ee.grid(row=0, column=2, columnspan=2)

        self.bb = Button(self.head, text='search', command=self.search)
        self.bb.grid(row=0, column=4, pady=10)
        
        self.b1 = Button(self.head, text='show all', command=self.show_table)
        self.b1.grid(row=0, column=5)

        self.g.trace_add('write', self.search)
        self.t.trace('w', self.search)


        #treeview
        co = self.head_list[:]
        self.cols = tuple(co)
        self.table = Treeview(self.w, columns=self.cols, show='headings')
        for c in self.cols:
            self.table.heading(c, text=c, command=lambda _c=c: self.sort_col(_c, False))
            self.table.column(c, anchor='center')
        self.table.grid(row=2, column=0, columnspan=4, rowspan=6, pady=10, padx=10)

        #Button

        self.b_frame = Frame(self.w)
        self.b_frame.grid(row=2,column=5,padx=10) 

        self.b1 = Button(self.b_frame, text='add', command=self.add_tab)
        self.b1.grid(row=1, column=0, padx=20)

        self.b1 = Button(self.b_frame, text='update', command=self.update_tab)
        self.b1.grid(row=2, column=0, padx=20)

        self.b1 = Button(self.b_frame, text='delete', command=self.delete)
        self.b1.grid(row=5, column=0, padx=20)
        
        self.w.mainloop()

app = tel_app(Tk())
```
